```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from datetime import datetime
import re
import json
from geopy.geocoders import Nominatim
import time

#Seed so that as we rerun, we keep getting the same results.  This ensures the analysis matches the results seen.
random.seed(1029384756) 

#When displaying a dataframe, the columns will not be truncated
pd.set_option('display.max_columns', None)  

parkingFacilities_file_string = "data/2017.02.05_Parking_Facilities.csv"
```

```python
parking_facilities_data = pd.read_csv(parkingFacilities_file_string)
```

```python
parking_facilities_data
```

```python
addressnumSeries = parking_facilities_data["address"]
streetSeries = parking_facilities_data["street"]

geolocator = Nominatim()

geoInfo = []
for i in range(len(parking_facilities_data)):
    addressToQuery = "{} {}, Baltimore, MD".format(str(addressnumSeries[i]).strip(), streetSeries[i].strip())
    print(addressToQuery)
    location = geolocator.geocode(addressToQuery)
    if location is not None:
        geoInfo.append((location.latitude, location.longitude))
    else:
        print("No Data")
        geoInfo.append(None)
    time.sleep(2)
```

Note that the below code was run because an exception was thrown in the above code (before we string-ified the addressnum to strip()).  However, the above code *should* work now.  It hasn't been tested so that we could comply with [Nomatim's Usage Policy](https://operations.osmfoundation.org/policies/nominatim/) and make as few queries as possible.
